# Test de Hipotesis

```python
sns.lmplot(x='IU', y='BILL_AMT', data=data, col='DEFAULT')
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

la distribucion de los grupos moros y no moroso es la misma

H<sup>alt<sup>

la distribucion de los grupos moros y no moroso no la misma

```python
moroso = data.query("DEFAULT == 1").BILL_AMT
```

```python
moroso
```

```python
no_moroso = data.query("DEFAULT == 0").BILL_AMT
```

```python
no_moroso
```

```python
resultado = ranksums(moroso, no_moroso)
```

```python
print("El p_value de nuestri tes es {}".format(resultado.pvalue))
```
